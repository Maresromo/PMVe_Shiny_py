**EN LA PARTE DE ABAJO (fecha1) TIENE QUE SER UNA FECHA EN LA QUE HAYA DATOS, POR QUE SI NO MARCARÁ UN ERROR**

```python
fecha1 = parse("2024-10-30")   #Año, mes , dia
fecha2 = fecha1 + pd.Timedelta("22D")

tmp= TB(config_file='config_esp.ini',device_name='est_Tlaltenango')
sensores = ["Temperatura","Humedad","Presion"]

# cafe es un nombre temporal, 
EST = pd.concat([tmp.get_df(key=sensor,start_datetime=fecha1,end_datetime=fecha2).resample('1min').mean() for sensor in sensores],axis=1)
```

```python
fecha1 = parse("30/10/2024 09:58", dayfirst=True)
fecha2 = parse("30/10/2024 10:06", dayfirst=True)

# Filtra el DataFrame por rango de fechas
EST_filtrada = EST.loc[fecha1:fecha2]
EST_filtrada
```
