```python
datos = [
    ("30/10/2024", "10:01", "10:06"),
    ("31/10/2024", "9:54", "10:02"),
    ("31/10/2024", "12:00", "12:04"),
    ("04/11/2024", "12:10", "12:19"),
    ("05/11/2024", "16:31", "16:41"),
    ("06/11/2024", "10:02", "10:10"),
    ("06/11/2024", "14:19", "14:35"),
    ("12/11/2024", "14:24", "14:43"),
    ("13/11/2024", "12:28", "12:36"),
    ("13/11/2024", "12:36", "12:45"),
    ("14/11/2024", "11:57", "12:06"),
    ("14/11/2024", "12:14", "12:23"),
    ("20/11/2024", "9:56", "10:02"),
    ("20/11/2024", "10:13", "10:18"),
]

# Crear las listas
fecha1_EST = [parse(f"{fila[0]} {fila[1]}", dayfirst=True) for fila in datos]
fecha2_EST = [parse(f"{fila[0]} {fila[2]}", dayfirst=True) for fila in datos]

# Mostrar los resultados
fecha2_EST[0]
```

```python
def filtro_EST(fecha1,fecha2):
    
    # Filtra el DataFrame por rango de fechas
    EST_filtrada = EST.loc[fecha1:fecha2]
    # Verificar si el valor de la primera fila en 'Humedad' supera en 10 o mÃ¡s al de la segunda fila
    if (EST_filtrada['Humedad'].iloc[0] - EST_filtrada['Humedad'].iloc[1]) >= 10:
        # Eliminar la primera fila
        EST_filtrada = EST_filtrada.iloc[1:]
    EST_filtrada = EST_filtrada.resample("D").mean()
    
    # Renombrar las columnas
    EST_filtrada = EST_filtrada.rename(columns={
        'Temperatura': 'Temperatura_prom',
        'Humedad': 'Humedad_prom',
        'Presion': 'Presion_prom'
        })
    return EST_filtrada.reset_index(drop=True)
```

```python
prubea = filtro_EST(fecha1_EST[1],fecha2_EST[1])
prubea
```

```python
resultados_EST = []
for f1, f2 in zip(fecha1_EST, fecha2_EST):
    df_filtrado = filtro_EST(f1, f2)
    resultados_EST.append(df_filtrado)

# Concatenar todos los DataFrames en uno solo
EST_concatenado = pd.concat(resultados_EST, ignore_index=True)
EST_concatenado
```

```python
def filtro_encuesta(fecha1,fecha2):
    
    # Filtra el DataFrame por rango de fechas
    encuesta_filtrada = encuesta.loc[fecha1:fecha2]
    
    return encuesta_filtrada
```

```python
prubea2 = filtro_encuesta(fecha1_encuesta[-1],fecha2_encuesta[-1])
prubea2
```

```python
resultados_encuesta = []
for f1, f2 in zip(fecha1_encuesta, fecha2_encuesta):
    df_filtrado = filtro_encuesta(f1, f2)
    resultados_encuesta.append(df_filtrado)

# Concatenar todos los DataFrames en uno solo
encuesta_concatenada = pd.concat(resultados_encuesta)
encuesta_concatenada.info()
```

```python
num_encuestados=[]
for df in resultados_encuesta:
    num_encuestados.append(df.shape[0])
num_encuestados
```

```python
# Lista con el nÃºmero de repeticiones
repeticiones = [10, 11, 3, 7, 4, 8, 11, 10, 9, 15, 6, 7, 10, 13]

# Verificar que el nÃºmero de filas en el DataFrame coincida con la longitud de la lista
if len(EST_concatenado) != len(repeticiones):
    raise ValueError("El nÃºmero de filas en el DataFrame no coincide con la longitud de la lista de repeticiones.")

# Crear una lista para almacenar los DataFrames repetidos
filas_repetidas = []

# Iterar por las filas y el nÃºmero de repeticiones
for i, n_reps in enumerate(repeticiones):
    fila_repetida = pd.DataFrame([EST_concatenado.iloc[i].values] * n_reps, columns=EST_concatenado.columns)
    filas_repetidas.append(fila_repetida)

# Concatenar todas las filas repetidas en un solo DataFrame
repetido = pd.concat(filas_repetidas, ignore_index=True)

repetido
```

```python
repetido.index = encuesta_concatenada.index
df_combinado = pd.concat([encuesta_concatenada, repetido], axis=1)

# Conservar el Ã­ndice de 'encuesta_concatenada'
df_combinado.index = encuesta_concatenada.index
df_combinado
```
