```python
from iertools.tb import *
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
```

```python
# pip install git+https://github.com/AltamarMx/iertools.git
```

```python
fecha1 = parse("2024-10-30")   #Año, mes , dia
fecha2 = fecha1 + pd.Timedelta("22D")

tmp= TB(config_file='config_encuesta.ini',device_name='Encuestas_Shiny')
parametros = ["clo","e","met","voto"]

encuesta = pd.concat([tmp.get_df(key=parametro,start_datetime=fecha1,end_datetime=fecha2) for parametro in parametros],axis=1)
```

```python
encuesta
```

```python
encuesta = encuesta.iloc[::-1]
encuesta
```

```python
fecha1 = parse("30/10/2024 10:01", dayfirst=True)
fecha2 = parse("30/10/2024 10:06", dayfirst=True)

# Filtra el DataFrame por rango de fechas
encuesta_filtrada = encuesta.loc[fecha1:fecha2]
encuesta_filtrada
```
